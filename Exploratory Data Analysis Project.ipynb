#### What are the locations with the fastest roller coasters, with a minimum of 10 coasters?

```python
df_roller_clean.head()
```

```python
df_roller_clean.groupby('Location')['Speed_Mph']\
                .agg(['mean','count'])\
                .sort_values(by='mean', ascending=False).head(20)
```

```python
ax=df_roller_clean.groupby('Location')['Speed_Mph']\
                .mean()\
                .sort_values(ascending=False).head(20)\
                .plot(kind='barh', figsize=(12,5), title='Averge Coast Speed by Location')

ax.invert_yaxis()
ax.set_xlabel('Average Coaster Speed')

plt.show()
```
